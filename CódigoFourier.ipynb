En este codigo se hara la expansión en series de fourier de la función x entre -pi y pi.
Se graficarán las sumatorias respectivas para los valores n=2,5,10,20,30

```python
%pylab inline
```

```python
x=linspace(-3*pi,3*pi,200)
```

### Para n=2

```python
def funcionX(n,x):
    return -2*((-1)**n)*sin(n*x)/n
```

```python
i=1
y2=linspace(0,0,200)
y2
while (i<=2):
    y2=y2+funcionX(i,x)
    i=i+1
```

```python
x1=linspace(-pi,pi,100)
y=x1
plt.plot(x,y2,c='blue', label='n=2')
plot(x1,y, c='green', label='f(x)=x')
legend(loc=0)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Expansión fourier de x entre -pi y pi para x con 2 terminos')
```

### Para n=5

```python
i=1
y5=linspace(0,0,200)
while(i<=5):
    y5=y5+funcionX(i,x)
    i=i+1
```

```python
plt.plot(x,y5,c='blue', label='n=2')
plot(x1,y, c='green', label='f(x)=x')
legend(loc=0)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Expansión fourier de x entre -pi y pi para x con 5 terminos')
```

### Comparacion

```python
plot(x,y2, c='green', label='n=2')
plot(x,y5, c='blue', label='n=5')
plot(x,y10, c='red', label='n=10')
plot(x,y20, c='purple', label='n=20')
plot(x,y30, c='black', label='n=30')
plot(x1,y, c='yellow', label='f(x)=x')
legend(loc=0)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Comparacion de las distintas aproximaciones a f(x)=x')
```
